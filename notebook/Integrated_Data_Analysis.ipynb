Data Cleaning

Changes the data type of the "Purchas_date" column from string to datetime

```python
df["Purchase_Date"] = pd.to_datetime(df["Purchase_Date"])
df.head(10)
```

Displays the result

```python
df.info()
```

Time Series Analysis

```python
#Monthly Sales

monthly_sales = df.groupby(df["Purchase_Date"].dt.to_period("M"))["Purchase_Amount"].sum()

monthly_sales.index = monthly_sales.index.astype(str)

plt.figure(figsize=(12,5))
monthly_sales.plot(marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Purchase Amount")
plt.xticks(rotation=45)
plt.savefig("../images/monthly_sales.png")
plt.show()
```

```python
#Quarterly Sales

quarterly_sales = df.groupby(df["Purchase_Date"].dt.to_period("Q"))["Purchase_Amount"].sum()

quarterly_sales.index = quarterly_sales.index.astype(str)

plt.figure(figsize=(10,5))
quarterly_sales.plot(kind="bar")
plt.title("Quarterly Sales")
plt.xlabel("Quarter")
plt.ylabel("Purchase Amount")
plt.savefig("../images/quaterly_sales.png")
plt.show()
```

Category Analysis

```python
#SALES BY CATEGORY

plt.figure(figsize=(10,5))

sns.barplot(
    x=df.groupby("Category")["Purchase_Amount"].sum().index,
    y=df.groupby("Category")["Purchase_Amount"].sum().values
)
plt.xticks(rotation=45)
plt.title("Sales by Category")
plt.savefig("../images/sales_by_category.png")

plt.show()
```

Relationship Analysis


```python
#AGE VS PURCHASE AMOUNT

plt.figure(figsize=(8,5))

sns.scatterplot(
    x="Age",
    y="Purchase_Amount",
    data=df
)

plt.title("Age vs Purchase Amount")
plt.savefig("../images/age_vs_sale.png")

plt.show()
```

```python
#RATING VS PURCHASE AMOUNT

plt.figure(figsize=(8,5))

sns.scatterplot(
    x="Rating",
    y="Purchase_Amount",
    data=df
)

plt.title("Rating vs Purchase Amount")
plt.savefig("../images/rating_vs_purchasing.png")


plt.show()
```

Distribution Analysis

```python
#PURCHASE AMOUNT

plt.figure(figsize=(8,5))

sns.histplot(
    df["Purchase_Amount"],
    bins=30,
    kde=True
)

plt.title("Distribution of Purchase Amount")
plt.savefig("../images/purchase_amount.png")


plt.show()
```

```python
#BOXPLOT

plt.figure(figsize=(8,5))

sns.boxplot(
    x=df["Purchase_Amount"]
)

plt.title("Purchase Amount Boxplot")
plt.savefig("../images/boxplot.png")


plt.show()
```

```python
#AGE DISRTRIBUTION

plt.figure(figsize=(8,5))

sns.histplot(
    df["Age"],
    bins=20,
    kde=True
)

plt.title("Age Distribution")
plt.savefig("../images/age_dist.png")

plt.show()
```

Correlation Heatmap

```python
# CORRELATION HEATMAP

plt.figure(figsize=(8,6))

numeric_df = df.select_dtypes(include=["number"])

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm",
    linewidths=0.5
)

plt.title("Correlation Heatmap")

plt.savefig("../images/correlation_heatmap.png")

plt.show()
```

KEY INSIGHTS

1. The monthly sales tend to show periods of higher and lower purchase activity, indicating seasonal fluctuations.
2. Certain product categories contribute more to total revenue than others.
3. A few cities account for a larger share of total purchases, while others contribute less.
4. The histogram suggests that purchase amounts are concentrated within a particular range, with relatively few very high-value purchases.
5. The box plot helps identify potential outliers in purchase amounts.
6. The scatter plots indicate whether age or customer ratings have any visible relationship with purchase amount.
7. The correlation heatmap shows the strength of relationships among the numerical variables in the dataset.
