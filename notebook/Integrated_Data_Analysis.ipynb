```python
#DATA CLEANING

df["Purchase_Date"] = pd.to_datetime(df["Purchase_Date"])
df.head(10)
```

```python
df.info()
```

```python
#TIME SERIES ANALYSIS

#Monthly Sales

monthly_sales = df.groupby(df["Purchase_Date"].dt.to_period("M"))["Purchase_Amount"].sum()

monthly_sales.index = monthly_sales.index.astype(str)

plt.figure(figsize=(12,5))
monthly_sales.plot(marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Purchase Amount")
plt.xticks(rotation=45)
plt.show()
```

```python
#Quarterly Sales

quarterly_sales = df.groupby(df["Purchase_Date"].dt.to_period("Q"))["Purchase_Amount"].sum()

quarterly_sales.index = quarterly_sales.index.astype(str)

plt.figure(figsize=(10,5))
quarterly_sales.plot(kind="bar")
plt.title("Quarterly Sales")
plt.xlabel("Quarter")
plt.ylabel("Purchase Amount")
plt.show()
```

```python
#CATEGORY ANALYSIS

#SALES BY CATEGORY

plt.figure(figsize=(10,5))

sns.barplot(
    x=df.groupby("Category")["Purchase_Amount"].sum().index,
    y=df.groupby("Category")["Purchase_Amount"].sum().values
)
plt.xticks(rotation=45)
plt.title("Sales by Category")
plt.show()
```

```python
#SALES BY CITY

city_sales = df.groupby("City")["Purchase_Amount"].sum().sort_values(ascending=False)

plt.figure(figsize=(12,5))

city_sales.plot(kind="bar")

plt.title("Sales by City")

plt.show()
```

```python
#PAYMENT METHOD DISTRIBUTION

df["Payment_Method"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(6,6)
)

plt.ylabel("")

plt.title("Payment Method Distribution")

plt.show()
```

```python
#RELATIONSHIP ANALYSIS

#AGE VS PURCHASE AMOUNT

plt.figure(figsize=(8,5))

sns.scatterplot(
    x="Age",
    y="Purchase_Amount",
    data=df
)

plt.title("Age vs Purchase Amount")

plt.show()
```

```python
#RATING VS PURCHASE AMOUNT

plt.figure(figsize=(8,5))

sns.scatterplot(
    x="Rating",
    y="Purchase_Amount",
    data=df
)

plt.title("Rating vs Purchase Amount")

plt.show()
```

```python
#DISTRIBUTION ANALYSI

#PURCHASE AMOUNT

plt.figure(figsize=(8,5))

sns.histplot(
    df["Purchase_Amount"],
    bins=30,
    kde=True
)

plt.title("Distribution of Purchase Amount")

plt.show()
```

```python
#BOXPLOT

plt.figure(figsize=(8,5))

sns.boxplot(
    x=df["Purchase_Amount"]
)

plt.title("Purchase Amount Boxplot")

plt.show()
```

```python
#AGE DISRTRIBUTION

plt.figure(figsize=(8,5))

sns.histplot(
    df["Age"],
    bins=20,
    kde=True
)

plt.title("Age Distribution")

plt.show()
```

```python
#CORRELATION HEATMAP

plt.figure(figsize=(8,5))

sns.histplot(
    df["Age"],
    bins=20,
    kde=True
)

plt.title("Age Distribution")

plt.show()
```
